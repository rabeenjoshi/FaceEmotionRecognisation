```python
import cv2
import sys
import numpy as np
DIRECTORY="C:\Python36\haarcascade_frontalface_default.xml"

video_capture = cv2.VideoCapture(0)
while True:
    Return, Frame = video_capture.read()
    face_classifier = cv2.CascadeClassifier(DIRECTORY)
    if Return is True:
        print("Started to read the video")
        grayscale_frame = cv2.cvtColor(Frame, cv2.COLOR_BGR2GRAY)
    else:
        continue
    faces = face_classifier.detectMultiScale(grayscale_frame, 1.3, 5)
    
    if not len(faces) > 0:
        print("No face detected")
    else:
        print("Detected frame")
        # draw box around faces
        for face in faces:
            (x,y,w,h) = face
            rectangle_Frame = cv2.rectangle(Frame,(x,y-30),(x+w,y+h+10),(255,0,0),2)
            newimg = rectangle_Frame[y:y+h,x:x+w]
            cropped_img = cv2.resize(newimg, (48,48), interpolation = cv2.INTER_CUBIC) / 255. #resizing for 
    
    cv2.imshow('Face Detection', cv2.resize(rectangle_Frame,None,fx=2,fy=2,interpolation = cv2.INTER_CUBIC))
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
print('STOPPPPED')
```
